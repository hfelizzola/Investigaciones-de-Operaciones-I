```python
import numpy as np
import pandas as pd
from gamspy import Container, Set, Parameter, Variable, Sum, Sense, Model, Equation, Options, Alias
```

```python
[i+1 for i in range(5)]
```

```python
# Modelo Matemático

# Container
m = Container()

# Set
i = Set(m , name='i', records=[i+1 for i in range(5)])
j = Alias(m, name="j", alias_with=i)

# Variable
x = Variable(m, name="x", domain=[i,j], type="positive")

# Función objetivo
objetivo = 100*x[1,2] + 30*x[1,3] + 20*x[2,3] + 10*x[3,4] + 15*x[4,2] + 60*x[3,5] + 50*x[4,5]

# Equation
nodo1 = Equation(m , name="nodo1")
nodo1[...] = 1 - x[1,2] - x[1,3] == 0

nodo2 = Equation(m , name="nodo2")
nodo2[...] = x[1,2] + x[4,2] - x[2,3] == 0

nodo3 = Equation(m , name="nodo3")
nodo3[...] = x[1,3] + x[2,3] - x[3,4] - x[3,5] == 0

nodo4 = Equation(m , name="nodo4")
nodo4[...] = x[3,4] - x[4,2] - x[4,5] == 0

nodo5 = Equation(m , name="nodo5")
nodo5[...] = x[3,5] + x[4,5] - 1 == 0

modelo = Model(m, objective=objetivo,
               problem="LP", equations=[nodo1, nodo2, nodo3, nodo4, nodo5],
               sense=Sense.MIN, name="modelo")
modelo.solve()
```
