La administración desea determinar un plan de distribución —número de cargas enviadas al mes de cada planta a cada almacén y de cada uno de éstos a cada tienda— de modo que se minimice el costo total de transporte.


**Modelo Matemático**

- Conjuntos:
  - $i$: Plantas, $I = \{1,2\}$
  - $j$: Almacenes, $J = \{1,2\}$
  - $k$: Tiendas, $K = \{1,2,3\}$
- Parámetros:
  - $C_{ij}$: Costo de transporte de $i \to j$
  - $B_{ij}$: Costo de transporte de $j \to k$
  - $S_i$: Capacidad del nodo $i$
  - $D_k$: Demanda del nodo $k$
- Variable de decisión:
  - $X_{ij}$: Cantidad a enviar de $i \to j$
  - $Y_{jk}$: Cantidad a enviar de $j \to k$
- Función Objetivo y Restricciones:

\begin{equation*}
\text{Min. } Z = \sum_{i=1}^{m}\sum_{j=1}^{n} C_{ij}X_{ij} + \sum_{j=1}^{m}\sum_{k=1}^{l} B_{jk}Y_{jk}
\end{equation*}
\begin{align*}
SA:\\
\sum_{j \in J} X_{ij} &= S_i && \forall i \in I\\
\sum_{i \in I} X_{ij} &= \sum_{k \in K} Y_{jk} && \forall j \in J\\
\sum_{j \in J} Y_{jk} &= D_k && \forall k \in K\\
X_{ij} &\ge 0, && \forall i \in I, \forall j \in J\\
Y_{jk} &\ge 0, && \forall j \in J, \forall k \in K\\
\end{align*}

```python
import numpy as np
import pandas as pd
from gamspy import Container, Set, Variable, Parameter, Equation, Model, Sense, Sum, Options
```

```python
m = Container()

# Conjuntos
planta = Set(m, name='planta', records=['P1','P2'])
almacen = Set(m, name='almacen', records=['A1','A2'])
tienda = Set(m, name='tienda', records=['T1','T2','T3'])

# Variable de decisión
x = Variable(m, name='x', type='Positive', domain=[planta,almacen])
y = Variable(m, name='y', type='Positive', domain=[almacen,tienda])

# Parametros
capacidad = Parameter(m, name='capacidad', domain=[planta], records=np.array([200,300]))
demanda = Parameter(m, name='demanda', domain=[tienda], records=np.array([150,200,150]))
costo_planta_almacen = Parameter(m, name='costo_planta_almacen',
                                 domain=[planta,almacen],
                                 records=np.array([[425,560],
                                                   [510,600]]))

costo_almacen_tienda = Parameter(m, name='costo_almacen_tienda',
                                 domain=[almacen,tienda],
                                 records=np.array([[470,505,490],
                                                   [390,410,440]]))

# Restricciones
capacidad_max = Equation(m, name='capacidad_max', domain=[planta])
capacidad_max[planta] = Sum(almacen, x[planta,almacen]) == capacidad[planta]

cons_flujo = Equation(m, name='cons_flujo', domain=[almacen])
cons_flujo[almacen] = Sum(planta, x[planta,almacen]) == Sum(tienda, y[almacen,tienda])

demanda_max = Equation(m, name='demanda_max', domain=[tienda])
demanda_max[tienda] = Sum(almacen, y[almacen,tienda]) == demanda[tienda]

costo_total = (Sum([planta,almacen], costo_planta_almacen[planta,almacen] * x[planta,almacen])
            + Sum([almacen,tienda], costo_almacen_tienda[almacen,tienda] * y[almacen,tienda]))

# Model
modelo_transporte = Model(m, name='modelo_transporte', objective=costo_total, sense=Sense.MIN, problem='LP', equations=[capacidad_max, cons_flujo, demanda_max])
modelo_transporte.solve()
```
